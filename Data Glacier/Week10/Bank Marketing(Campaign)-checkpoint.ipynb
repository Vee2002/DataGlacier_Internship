#### i) Univariate Analysis

```python
# Function to plot pie charts

def pie_charts(column_name,colors,title):
    plt.figure(figsize=(14,6))
    column_counts = data[column_name].value_counts()
    plt.pie(column_counts,labels=column_counts.index,autopct='%1.1f%%',colors=list(colors),textprops={'fontsize':12,'color':'white'})
    plt.legend(column_counts.index,title=title,loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()

# Funtion to plot bar charts
# Passing kind=kind cause of preference for horizontal bar charts in some plots
def bar_charts(column_name,kind,color,rotation_degree,title_name,xlabel,ylabel):
    plt.figure(figsize=(14,6))
    column_counts = data[column_name].value_counts()
    column_counts.plot(kind=kind,color=list(color))
    plt.title(title_name)
    plt.xticks(rotation=rotation_degree)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

#### ii) Bivariate Analysis

Creating a function to plot 2 categories by grouping.Example, job by purchased deposit, education by purchased deposit etc.

```python
def group_categories(column_one,column_two,color,xlabel,ylabel,title):
    # Grouping the data
    grouped_data = data.groupby([column_one,column_two],observed=True).size().reset_index(name='count')
    # Reshaping the data
    pivot_data = grouped_data.pivot(index=column_one,columns=column_two,values='count')
    
    # Grouped bar plot
    pivot_data.plot(kind='bar',figsize=(15,8),stacked=False,color=color)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

```python
# Job by purchased deposit
group_categories('job','purchased_deposit',['#2f6690','#8093f1'],'Job Categories','Counts','Grouped Bar Plot for Job and Purchased Deposit')
```
