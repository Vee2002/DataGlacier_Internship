Creating a function to plot 2 categories by grouping.Example, job by purchased deposit, education by purchased deposit etc.

```python
def group_categories(column_one,column_two,color,xlabel,ylabel,title):
    # Grouping the data
    grouped_data = data.groupby([column_one,column_two],observed=True).size().reset_index(name='count')
    # Reshaping the data
    pivot_data = grouped_data.pivot(index=column_one,columns=column_two,values='count')
    
    # Grouped bar plot
    pivot_data.plot(kind='bar',figsize=(15,8),stacked=False,color=color)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

I've one hot encoded most of the categorical columns apart from month and day_of_the week.These two columns represent a cyclic nature and ordinal encoding is not the best option because it will assign1 to January and 12 to December suggesting that they are far from each other which is false.It may give more importance to January.

One hot encoding on the other hand will lead to so many columns or features.So to solve this, since models cannot take Monday and October, we'll perform cyclic encoding which retains the concept of a cycle 

#### Cyclic Encoding

```python
data['month'].value_counts()
```

#### Changing month

```python
# Month order
month_order = ['mar','apr','may','jun','jul','aug','sep','oct','nov','dec']

# Converting month to numeric index
data['month_num'] = data['month'].map(lambda x: month_order.index(x) + 1).astype(float)

# Applying cyclic transformation with months = 10 instead of 12 as we have 10 months in our dataset
data["month_sin"] = np.sin(2 * np.pi * data["month_num"] / 10)
data["month_cos"] = np.cos(2 * np.pi * data["month_num"] / 10)

data.drop(columns=["month_num"], inplace=True)
```

#### Changing week

```python
data['day_of_week'].value_counts()
```

```python
# Month order
week_order = ['mon','tue','wed','thu','fri']

# Converting month to numeric index
data['week_num'] = data['day_of_week'].map(lambda x: week_order.index(x) + 1).astype(float)

# Applying cyclic transformation with months = 10 instead of 12 as we have 10 months in our dataset
data["week_sin"] = np.sin(2 * np.pi * data["week_num"] / 10)
data["week_cos"] = np.cos(2 * np.pi * data["week_num"] / 10)

data.drop(columns=["week_num"], inplace=True)
```

#### Normalization

Normalization is applied on features with different scales and it scales numerical features to a range of [0,1] or [-1,1].

Normalization is necessary whenu sing Distance-Based Models such as SVM, KNN and Neural Networks.

And not necessary for Tree-Based Models such as Random Forest, XGBoost and LightGBM) as they are scale invariant.

For modelling, the requirements suggest having a baseline model,ensemble model and a boosting one.

In this project I will use Logistic Regression as my baseline model, Random Forest as my ensemble model and XGBoost for boosting.Since I'll use Logistic Regression as my baseline and it is sensitive to feature scales and benefits from features being on the same scale, normalization is important

## MODELLING

```python
# Splitting the orginal data into train and 

# Excluding target column
original_X = data.drop(columns=['purchased_deposit'])
original_y = data['purchased_deposit']

original_X_train,original_X_test,original_y_train,original_y_test = train_test_split(original_X,original_y,test_size=0.2,random_state=42)
```

```python
# Initializing the model
lr = LogisticRegression()

# Fitting the model
lr.fit(original_X_train,original_y_train)

# Predicting
original_y_pred = lr.predict(original_X_test)

# Evaluation Metrics
accuracy = accuracy_score(original_y_test, original_y_pred)
precision = precision_score(original_y_test,original_y_pred,average="binary")
recall = recall_score(original_y_test,original_y_pred,average="binary")
f1 = f1_score(original_y_test,original_y_pred,average="binary")

# Printing Metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

### EVALUATION

Accuracy - Measures overall accuracy

Precision - When False Positives matter the most

Recall - When False Negatives are more important

F1-Score - Best used when both precision and recall are important

For our project, Precision(False positives) means predicting that the customer will purchase the term deposit and they do not.This leads to waste of marketing resources which we do not want to be the case.

While Recall(False Negatives) means predicting that the customer will not purchase the term deposit and they actually do,leading to missed opportunity for revenue which we also do not want to be the case.

To cater for both situations, we will use F1-Score.

Logistic Regression has an F1-Score of 48.51%.

Radom Forest has an improved the F1-Score which is 55.11%.

And finally XGBoost with the highest F1-Score of 58.61%.Thus this our best performing model
